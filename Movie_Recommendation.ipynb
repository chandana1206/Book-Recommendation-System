```python
# Convert genres, cast, and crew Columns into lists
import ast

def extract_genres(x):
    genres = []
    if x:
        for i in ast.literal_eval(x):
            genres.append(i['name'])
    return genres

def extract_cast(x):
    cast = []
    if x:
        for i in ast.literal_eval(x)[:3]:  # Top 3 actors
            cast.append(i['name'])
    return cast

def extract_director(x):
    directors = []
    if x:
        for i in ast.literal_eval(x):
            if i['job'] == 'Director':
                directors.append(i['name'])
    return directors
```

```python
movies_df['genres'] = movies_df['genres'].apply(extract_genres)
movies_df['cast'] = movies_df['cast'].apply(extract_cast)
movies_df['crew'] = movies_df['crew'].apply(extract_director)
```

```python
movies_df = movies_df.rename(columns={'crew': 'director'})
movies_df = movies_df.rename(columns={'vote_average': 'votes'})
```

```python
# Convert 'genres' and 'cast' columns to strings if they're in list format
movies_df['genres'] = movies_df['genres'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
movies_df['cast'] = movies_df['cast'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
movies_df['director'] = movies_df['director'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)
```

```python
# Convert all columns with string data to lowercase
movies_df = movies_df.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
movies_df.head()
```

```python
# Combine 'overview', 'genres', 'cast', and 'director' into a single 'combined' column
movies_df['tags'] = (
    movies_df['overview'].fillna('') + ' ' +
    movies_df['genres'].apply(lambda x: ' '.join(x)) + ' ' +
    movies_df['cast'].apply(lambda x: ' '.join(x)) + ' ' +
    movies_df['director'].apply(lambda x: ' '.join(x))
)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
import string

def lemmatization(tags):
    words = []
    for word in tags.split():
        # Lemmatize the word and remove punctuation
        word = word.strip(string.punctuation)
        words.append(lemmatizer.lemmatize(word))
    return ' '.join(words)

movies_df['tags'] = movies_df['tags'].apply(lemmatization)
```

```python
# Term Frequency-Inverse Document Frequency 
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(stop_words='english') #removes common English words (like "the", "is", "in", etc.)

# Fit and transform the 'tags' column into numerical vectors
tfidf_matrix = vectorizer.fit_transform(movies_df['tags']).toarray()

# to see how many features weâ€™ve got
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def get_similar_movies(title, cosine_sim=cosine_sim):
    
    title = title.lower()

    # Find the index of the movie in the dataset, if it exists
    if title in movies_df['title'].str.lower().values:
        idx = movies_df[movies_df['title'].str.lower() == title].index[0]
    else:
        print("Movie not found in the dataset.")
        return
    
    # Get the similarity scores of all movies with the input movie
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort the movies based on similarity scores (highest first)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get the top 5 most similar movies (skip the first one, because it's the movie itself)
    top_movies = sim_scores[1:6]
    
    # Extract movie titles of the top 5 most similar movies
    for i in top_movies:
        print( movies_df['title'].iloc[i[0]])
```

```python
def get_top_movies_with(genre=None, actor=None, director=None, num_recs=10):
    # Start with all movies
    filtered_movies = movies_df

    # Filter by genre if provided
    if genre:
        filtered_movies = filtered_movies[filtered_movies['genres'].str.contains(genre.lower())]

    # Filter by actor if provided
    if actor:
        filtered_movies = filtered_movies[filtered_movies['cast'].str.contains(actor.lower())]

    # Filter by director if provided
    if director:
        filtered_movies = filtered_movies[filtered_movies['director'].str.contains(director.lower())]

    # If no movies are found after all filters, return an empty list
    if filtered_movies.shape[0] == 0:
        return "No Movies Found"

    # Sort by ratings in descending order
    filtered_movies = filtered_movies.sort_values(by='votes', ascending=False)

    # Return the top N movie titles
    return filtered_movies['title'].head(num_recs)
```
